```python
import datetime
import json, requests, sys, time
from pprint import pprint
import os
import csv

def fetch_api_url(url):
    r = requests.get(url)
    #time.sleep(0.5) ## necessary to make sure that the output is fully fetched
    content_url = r.content
    # Put content_url in json variable and return it
    json_val = {}
    try:
        json_val = json.loads(content_url)
    except:
        pass
    return json_val

query1 = "https://www.peeringdb.com/api/ixpfx"
list_IXPs = fetch_api_url(query1)

print ('Num IXPs in PeeringDB = ', len(list_IXPs['data']))

'''
PCH=[]
id=[]
check={'error': 'No IXP Data found. Please check your query parameters and again later.'}
with open('D://third/python/temp.csv','w',encoding='GB18030',newline='')as csvfile:
    writer=csv.writer(csvfile)
    writer.writerow(['id','protocol','prefix'])
    i=2736
    for i in range(2500):
        query_temp='https://www.pch.net/api/ixp/subnets/'+str(i)
        temp=fetch_api_url(query_temp)
        print(temp)
        if(temp[0]!=check):
            for k in range(len(temp)):
                if(temp[k]['id'] not in id):
                    id_PCH=temp[k]['id']
                    prefix_PCH=temp[k]['subnet']
                    protocol_PCH=temp[k]['version']
                    PCH.append({'id_PCH':id_PCH,'prefix_PCH':prefix_PCH,'protocol_PCH':protocol_PCH})
                    id.append(id_PCH)
                    writer.writerow([id_PCH,protocol_PCH,prefix_PCH])
        print(i)
'''

file=open('temp.csv')
file.readline()

PCH=[]

for line in file:
    items=line.strip().split(',')
    id_PCH=str(items[0])
    protocol_PCH=str(items[1])
    prefix_PCH=str(items[2])
    PCH.append({'id_PCH':id_PCH,'protocol_PCH':protocol_PCH,'prefix_PCH':prefix_PCH})
```

```python
PDB=[]
for i in range(len(list_IXPs['data'])):
    id_PDB=list_IXPs['data'][i]['ixlan_id']
    prefix_PDB=list_IXPs['data'][i]['prefix']
    protocol_PDB=list_IXPs['data'][i]['protocol']
    PDB.append({'id_PDB':id_PDB,'prefix_PDB':prefix_PDB,'protocol_PDB':protocol_PDB})
```

```python
mapping={}
for i in range(len(PDB)):
    for j in range(len(PCH)):
        if(PDB[i]['prefix_PDB']==PCH[j]['prefix_PCH']):
            mapping[PDB[i]['id_PDB']]=PCH[j]['id_PCH']
```

```python
file=open('IXP.csv')
file.readline()

data=[]

for line in file:
    items=line.strip().split(',')
    Country=str(items[0])
    IXPDB=str(items[1])
    PCH=str(items[2])
    PDB=str(items[3])
    name=str(items[4])
    city=str(items[5])
    region=str(items[6])
    creation=str(items[7])
    url=str(items[8])
    stats=str(items[9])
    check=str(items[10])
    checkdate=str(items[11])
    #print(check)
    data.append({'Country':Country,'IXPDB':IXPDB,'PCH':PCH,'PDB':PDB,'name':name,'city':city,'region':region,'creation':creation,'url':url,'stats':stats,'check':check,'checkdate':checkdate})
```

```python
keys=[]
values=[]
for i in mapping.keys():
    keys.append(i)
    values.append(mapping[i])
```

```python
#keys
```

```python
for i in range(len(data)):
    for key in keys:
        if(str(key)==data[i]['PDB']):
            data[i]['PCH']=mapping[key]
            #print(data[i]['PCH'])
```
